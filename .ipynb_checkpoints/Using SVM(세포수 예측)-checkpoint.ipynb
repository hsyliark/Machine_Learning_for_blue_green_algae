# Support Vector Machine

## Data reading

```python
import pandas as pd
blue_algae = pd.read_csv('D:/Workplace/Machine_Learning_for_blue_green_algae/전처리 후 데이터/predict_blue_algae.csv', 
                         sep=',', index_col='date')
blue_algae.head(5)
```

```python
blue_algae.info()
```

## Making dependent and independent

```python
dataset = blue_algae.values
X = dataset[:,0:20]
y = dataset[:,20]

# normalization
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
X_std
```

## SVM regression modeling

```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score

rmse_result = []
rsquare_result = []

for step in range(30001):
    X_std_train, X_std_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2)
    svm_ploy_reg = SVR(kernel='rbf', C=100, gamma='auto', epsilon=.1)
    model = svm_ploy_reg.fit(X_std_train, y_train)
    y_pred_svm = model.predict(X_std_test)
    y_bar = np.mean(y_test)
    rmse = np.sqrt(np.mean(np.square(y_pred_svm - y_test)))
    SST = np.sum(np.square(y_test - y_bar))
    SSReg = np.sum(np.square(y_pred_svm - y_bar))
    rsquare = SSReg/SST  
    rmse_result.append(rmse)
    rsquare_result.append(rsquare)
    if step % 1000 == 0:
        print("Epoch: %.f, RMSE: %.04f, RSQUARE: %.04f" % (step, rmse, rsquare))
```
