# Deep Neural Network

## Data reading

```python
import pandas as pd
blue_algae = pd.read_csv('D:/Workplace/Machine_Learning_for_blue_green_algae/전처리 후 데이터/predict_blue_algae.csv', 
                         sep=',', index_col='date')
blue_algae.head(5)
```

```python
blue_algae.info()
```

## Making dependent and independent

```python
dataset = blue_algae.values
X = dataset[:,0:20]
y = dataset[:,20]

# normalization
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
X_std
```

## DNN

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

model = Sequential()
model.add(Dense(3000, input_dim=20, activation='relu'))
model.add(Dense(1250, activation='relu'))
model.add(Dense(600, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(400, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse','mae'])

history = model.fit(X_std, y, validation_split=0.3, epochs=4000, batch_size=100)
```

```python
y_prediction = model.predict(X).flatten() # 검증 데이터에 대한 예측값, flatten : 데이터를 1차원으로 바꿈
for i in range(len(y)):
    label = y[i]
    prediction = y_prediction[i]
    print("실제세포수: {:.3f}, 예상세포수: {:.3f}".format(label, prediction))
```
